# TEST OF LINEAR REGRESSION MODULE

In this notebook we test the module of  **linear regression** that we created. We will test whether our model can fit a y = 5x+100 line. If we can fit the line then we can positively conclude that our module is working.

**Add the directory of the m_learn package to the system path**

```python
from config import append_path
append_path('../../')
```

**Import necessary packages and modules**

```python
import pandas as pd
import numpy as np
from m_learn.linear_model import linear_regression
from m_learn.utility.data_preprocessing import get_xy_data
import matplotlib.pyplot as plt
%matplotlib inline 
%config InlineBackend.figure_format = 'retina'
```

## 1. Test whether the model can fit y = 5x+100

### 1.1 Create and fit the model

**create the x and y data**

```python
x = np.linspace(0,100,51)
y = 5*np.copy(x)+100
```

**store the points in pandas dataframe**

```python
data = pd.DataFrame({'x':x,"y": y})
data.head()
```

**feature and output of the model**

```python
features = ['x']
output = 'y'
```

**extracting the feature matrix (with the column of intercept) and output vector**

```python
(X, y) = get_xy_data(data, features, output)
```

**stepsize and tolerance of the model**

```python
step_size = 4e-6
tolerance = 1e-1
```

**create and fit a linear regression model using the data**

```python
model = linear_regression()
model.fit(X, y, step_size, tolerance, verbose=True)
```

### 1.2 Verify the fitted model

**Inspect the model coefficients**

```python
model.coefficients
```

**Compute predictions using the feature matrix**

```python
predictions = model.predict(X)
```

#### Line plots of data and the model predictions

```python
plt.plot(x,y, c = 'k', linewidth = 2)
plt.scatter(x, predictions, c = 'r')
plt.title('Regression Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```
