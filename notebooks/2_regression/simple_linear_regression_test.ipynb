# TEST OF SIMPLE LINEAR REGRESSION MODULE

In this notebook we test the module of **simple linear regression** that we created. We will test whether our mini model can fit a **y = x** & **y = 5x + 100** lines. If we can fit these lines we can positively conclude that our simple linear regression module working is working.

**Add the directory of m_learn package to the system path**

```python
from config import append_path
append_path('../../')
```

**Import packages and modules**

```python
import numpy as np
import pandas as pd
from m_learn.linear_model import simple_linear_regression
import matplotlib.pyplot as plt
%matplotlib inline 
%config InlineBackend.figure_format = 'retina'
```

## 1. Testing whether the model can fit y = x

**Create linearly spaced xy data for y = x line**

```python
x = np.linspace(0,100,51)
y = np.copy(x)
```

**Create a simple linear regression model and fit the model using the data**

```python
model = simple_linear_regression()
model.fit(x,y)
```

**Verify the slope & intercept**

```python
model.slope
```

```python
model.intercept
```

**Compute predictions using x values**

```python
predictions = model.predict(x)
```

#### Line plots of the data and the model predictions

```python
plt.plot(x,y, c = 'k')
plt.scatter(x, predictions, c='r')
plt.title('Regression Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```

## 2. Testing whether the model can fit y = 5x+100

**Create the xy data for y = 5x+100 line and fit a simple linear regression model using the data**

```python
x = np.linspace(0,100,51)
y = 5*np.copy(x)+100
model = simple_linear_regression()
model.fit(x,y)
predictions = model.predict(x)
```

**Verify the slope & intercept**

```python
model.slope
```

```python
model.intercept
```

#### Line plots of the data and the model predictions

```python
plt.plot(x,y, c = 'k')
plt.scatter(x, predictions, c = 'r')
plt.title('Regression Plot')
plt.xlabel('x')
plt.ylabel('y')
plt.grid()
plt.show()
```
