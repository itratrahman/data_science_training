#  1 Bar Plot

Bar graph is mainly used to compare between two categories or groups. For example you want to show the comparison between employees and their incomes.

## 1.1 Column Bar Chart

```python
## Basic Plot

plt.figure(figsize=(10,10))
id1 = np.arange(1,10)
salary = np.arange(20,110,10)
plt.bar(id1,salary)
plt.xlabel('Employee ID')
plt.ylabel('Salary')
plt.show()
```

**We can style the graph using the following functions:**

* **plt.title( )** for specifying title of your plot.
* **plt.xlabel( )** for labeling x-axis.
* **plt.ylabel( )** for labeling y-axis.
* **color** = option in plt.bar( ) for defining color of bars.

```python
id1 = np.arange(1,10)
salary = np.arange(20,110,10)
plt.figure(figsize=(8,5))                # Setting the figure size
ax = plt.axes()
ax.set_facecolor("#ECF0F1")              # Setting the background color by specifying the HEX Code
plt.bar(id1,salary,color = '#FFA726')
plt.xlabel(r'$Employee $ $ ID$')
plt.ylabel(r'$Salary$')
plt.show()
```

```python
#Plotting multiple sets of data

x1= [1,3,5,7]
x2=[2,4,6,8]
y1 = [7,7,7,7]
y2= [17,18,29,40]
plt.figure(figsize=(8,6))
ax = plt.axes()
ax.set_facecolor("white")
plt.bar(x1,y1,label = "First",color = '#42B300')     # First set of data
plt.bar(x2,y2,label = "Second",color = '#94E413')   # Second set of data
plt.xlabel('$X$')
plt.ylabel('$Y$')
plt.title ('$Bar $ $ Chart$')
plt.legend()
plt.show()
```

### How to show values and label at the top of the graph

```python
barplot = plt.bar(id1, salary)
for bar in barplot:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval, int(yval), va='bottom') #va: vertical alignment y positional argument
    
plt.title(" Bar graph example")
plt.xlabel('Employees')
plt.ylabel("Salary")
```

**Code Explanation**

1. **.get_height()** returns height of rectangle of each bar which is basically a value of y-axis
2. **plt.text()** is used to place text on the graph
3. **get_x() and get_width()** is used to find the value of the x-axis

## 1.2 Horizontal Bar Chart

```python
plt.figure(figsize=(10,10))
y,x = zip(*sorted(zip(salary,id1)))
plt.barh(x,y)
plt.title("Horizontal Bar graph example")
plt.xlabel("Employees")
plt.ylabel('Salary')
```

We can arrange the bars based on the values. We need to combine both the lists and the sort them based on value of list. **zip()** function is used to combine items of lists x and y. **sorted()** function sort them based on y. Then we split and store it in x and y lists.

## 1.3 Stacked Vertical Bars

```python
plt.style.use('seaborn-darkgrid')
x1= ['Sachin','Saurav','Rahul','Robin']
y1= [17,18,29,40]
y2 = [20,21,22,23]
plt.figure(figsize=(5,7))
plt.bar(x1,y1,label = "Batting Average",width = 0.5,color = '#FF6F00')
plt.bar(x1,y2,label = "Bowling Average",width = 0.5 ,bottom = y1 , color = '#FFB300')
plt.xlabel('$X$')
plt.ylabel('$Y$')
plt.title ('$Bar $ $ Chart$')
plt.legend()
plt.show()
```

## 1.4 Grouped Bar Charts

```python
# Grouped Bar Chart
plt.figure(figsize=(7,9))
# set width of bar
barWidth = 0.25
# set height of bar
y1= np.array([17,18,29,40])
y2 =np.array([20,21,22,23])
y3 =np.array([5,9,11,12])
# Set position of bar on X axis
pos1 = np.arange(len(y1))
pos2 = [x + barWidth for x in pos1]
pos3 = [x + barWidth for x in pos2]
# Make the plot
plt.bar(pos1, y1, color='#FBC02D', width=barWidth, label='Batting')
plt.bar(pos2, y2, color='#F57F17', width=barWidth, label='Bowling')
plt.bar(pos3, y3, color='#E65100', width=barWidth, label='Fielding')
# Add xticks on the middle of the group bars
plt.xlabel('Assignee', fontweight='bold')
plt.ylabel('Number of Tickets', fontweight='bold')
plt.xticks([i + barWidth for i in range(len(y1))], ['Sachin', 'Saurav', 'Rahul', 'Robin'])

# Create legend & Show graphic
plt.legend()
plt.show()
np.arange(len(y1))
```

```python
plt.style.use('seaborn-darkgrid')
x1= ['Saurav','Sachin','Ravi','Minil']
y1= [17,18,29,40]
y2 = [20,21,22,23]
plt.figure(figsize=(8,5))
plt.barh(x1,y1,label = "Batting Tickets",color = '#FF6F00')
plt.barh(x1,y2,label = "Bowling Tickets", left = y1 , color = '#FFB300')
plt.xlabel('$X$')
plt.ylabel('$Y$')
plt.title ('$Bar $ $ Chart$')
plt.legend()
plt.show()
```

### Displaying values in Bar Charts

```python
# Displaying values in the stacked vertical bars using plt.text()

plt.style.use('seaborn-darkgrid')
x1= ['Saurav','Sachin','Rahul','Robin']
y1= [17,18,29,40]
y2 = [20,21,22,23]
plt.figure(figsize=(5,7))
plt.bar(x1,y1,label = "Batting Average",width = 0.5,color = '#FF6F00')
plt.bar(x1,y2,label = "Bowling Average",width = 0.5 ,bottom = y1 , color = '#FFB300')
plt.xlabel('$X$')
plt.ylabel('$Y$')
plt.title ('$Bar $ $ Chart$')
for x,y in zip(x1,y1):
    plt.text(x, y-10, '%d' % y, ha='center' , va= 'bottom')
for x,y,z in zip(x1,y2,y1):
    plt.text(x, y+z-10, '%d' % y, ha='center' , va= 'bottom')
plt.legend()
plt.show()
```

```python
# Displaying values at the top of the Grouped Bar Chart using plt.text()

plt.figure(figsize=(7,9))
# set width of bar
barWidth = 0.25
# set height of bar
y1= np.array([17,18,29,40])
y2 =np.array([20,21,22,23])
y3 =np.array([5,9,11,12])
# Set position of bar on X axis
pos1 = np.arange(len(y1))
pos2 = [x + barWidth for x in pos1]
pos3 = [x + barWidth for x in pos2]
# Make the plot
plt.bar(pos1, y1, color='#FBC02D', width=barWidth, label='Open')
plt.bar(pos2, y2, color='#F57F17', width=barWidth, label='Closed')
plt.bar(pos3, y3, color='#E65100', width=barWidth, label='Cancelled')

# Add xticks on the middle of the group bars
plt.xlabel('Assignee', fontweight='bold')
plt.ylabel('Number of Tickets', fontweight='bold')
plt.xticks([i + barWidth for i in range(len(y1))], ['Asif', 'Basit', 'Ravi', 'Minil'])
for x,y in zip(pos1,y1):
    plt.text(x, y, '%d' % y, ha='center' , va= 'bottom')
for x,y in zip(pos2,y2):
    plt.text(x, y, '%d' % y, ha='center' , va= 'bottom')
for x,y in zip(pos3,y3):
    plt.text(x, y, '%d' % y, ha='center' , va= 'bottom')
plt.title ('$Grouped $ $ Bar $ $ Chart$')

# Create legend & Show graphic
plt.legend()
plt.show()
```

# 4 Pie Chart

A pie chart is a circular graph which splits data into slices to show numerical proportion of each category. If you are showing percentages, all of them should add to 100%.

```python
Goals = [20, 12, 11, 4, 3]
players = ['Ronaldo', 'Messi', 'Suarez', 'Neymar', 'Salah', ]
comp = pd.DataFrame({"Goals" : Goals, "players" : players})
ax = comp.plot(y="Goals", kind="pie", labels = comp["players"], autopct = '%1.0f%%', legend=False, title='No of Goals scored')

# Hide y-axis label
ax.set(ylabel='')
```

## 4.1 Display percentage and actual value in Pie charts

```python
# Display percentage in Pie Chart using autopct='%1.1f%%'
plt.figure(figsize=(8,8))
area = [48 , 30 , 20 , 15]
labels = ['Low' , 'Medium' , 'High' , 'Critical']
colors = ['#7CB342','#C0CA33','#FFB300','#F57C00']
plt.pie (area , labels= labels , colors= colors , startangle=45 , shadow='true', autopct='%1.1f%%')
plt.show()
```

```python
plt.figure(figsize=(8,8))
area = [48 , 30 , 20 , 15]
total = np.sum(area)
labels = ['Low' , 'Medium' , 'High' , 'Critical']
def val_per(x):
    return '{:.2f}%\n({:.0f})'.format(x, total*x/100)
colors = ['#7CB342','#C0CA33','#FFB300','#F57C00']
plt.pie (area , labels= labels , colors= colors , startangle=45 , shadow='true', autopct=val_per)
plt.show()
```

# 5 Donut Plot

```python
plt.figure(figsize=(9,9))
area = [48 , 30 , 20 , 15]
labels = ['Low' , 'Medium' , 'High' , 'Critical']
colors = ['#8BC34A','#D4E157','#FFB300','#FF7043']
plt.pie (area , labels= labels , colors= colors , startangle=45)
my_circle=plt.Circle( (0,0), 0.7, color='white') # Adding circle at the centre
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.show()
```

## 5.1 Exploding slice in Donut chart

```python
plt.figure(figsize=(9,9))
area = [48 , 30 , 20 , 15]
labels = ['Low' , 'Medium' , 'High' , 'Critical']
colors = ['#8BC34A','#D4E157','#FFB300','#FF7043']
plt.pie (area , labels= labels , colors= colors , startangle=45 , explode=[0.03,0.03 , 0.03,0.03])
my_circle=plt.Circle( (0,0), 0.7, color='white') # Adding circle at the centre
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.show()
```

## 5.2 Displaying percentage and actual values in Donut chart

```python
plt.figure(figsize=(9,9))
area = [48 , 30 , 20 , 15]
labels = ['Low' , 'Medium' , 'High' , 'Critical']
colors = ['#8BC34A','#D4E157','#FFB300','#FF7043']
plt.pie (area , labels= labels , colors= colors , startangle=45 , autopct='%1.1f%%', pctdistance=0.8)
my_circle=plt.Circle( (0,0), 0.7, color='white') # Adding circle at the centre
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.show()
```

```python
plt.figure(figsize=(9,9))
area = [48 , 30 , 20 , 15]
total = np.sum(area)
def val_per(x):
    return '{:.2f}%\n({:.0f})'.format(x, total*x/100)
labels = ['Low' , 'Medium' , 'High' , 'Critical']
colors = ['#8BC34A','#D4E157','#FFB300','#FF7043']
plt.pie (area , labels= labels , colors= colors , startangle=45 , autopct=val_per, pctdistance=0.8)
my_circle=plt.Circle( (0,0), 0.7, color='white') # Adding circle at the centre
p=plt.gcf()
p.gca().add_artist(my_circle)
plt.show()
```

# 8 Box Plot

A Box plot also known as a whisker plot displays a summary of a set of data containing the minimum, first quartile, median, third quartile, and maximum. Box plots are widespread in descriptive statistics, they allow to quicklyexplore one or more datasets. 

In a box plot, we draw a box from the first quartile to the third quartile. A vertical line goes through the box at the median. The whiskers go from each quartile to the minimum or maximum.

```python
# Set the numpy seed for results reproducibility 
np.random.seed(23)

# Generate data
data = np.random.normal(0, 5, size=400)

# create a boxplot
plt.boxplot(data)
plt.show()
```

## 8.1 Horizontal box plot

```python
np.random.seed(23)
data = np.random.normal(3, 6, size=1000)
plt.boxplot(data, vert=False)
plt.show()
```

## 8.2 Multiple box plots

```python
np.random.seed(23)

data_1 = np.random.normal(50, 40, 200)
data_2 = np.random.normal(60, 30, 200)
data_3 = np.random.normal(70, 20, 200)
data_4 = np.random.normal(80, 10, 200)
data_5 = np.random.normal(90, 35, 200)

data = [data_1, data_2, data_3, data_4, data_5]
plt.figure(figsize=(10, 7))
plt.boxplot(data)

plt.show()
```

## 8.3 Box plot labels

```python
np.random.seed(23)
 
data_1 = np.random.normal(50, 40, 200)
data_2 = np.random.normal(60, 30, 200)
data_3 = np.random.normal(70, 20, 200)
data_4 = np.random.normal(80, 10, 200)
data_5 = np.random.normal(90, 35, 200)

data = [data_1, data_2, data_3, data_4, data_5]

plt.figure(figsize =(10, 7))

labels = ['first', 'second', 'third', 'fourth','fifth']
plt.boxplot(data, labels=labels)

plt.ylabel('Values')
plt.xlabel('Data sets')
plt.title('Multiple box plot example', fontsize=16)

plt.show()
```

# 9 Violin Plot

# 10 Heatmap

A heatmap is a graph that extensively uses color for data visualization. The colors depend on several independant variables. there are two different kinds of heatmap: a cluster heatmap (a cell-matrix of different colors) and a spatial heatmap (it has no cells; the variable variation is considered continuous).
